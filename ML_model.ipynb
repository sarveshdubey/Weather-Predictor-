```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
#Reading CSV file 
df = pd.read_csv("ML_model.csv")

df
```

```python
#Encoding into numbers
Numerics = LabelEncoder()
```

```python
#Dropping target variable(s) and making them as new frame(s)
inputs = df.drop('Rain',axis='columns')
target = df['Rain']
target
```

```python
inputs['Description_n'] = Numerics.fit_transform(inputs['Description'])
inputs

    
```

```python
inputs_n = inputs.drop(['Description','City','Country','Longitude','Lattitude','Sunset','Sunrise'],axis='columns')
inputs_n
```

```python
Classifier = GaussianNB()
Classifier.fit(inputs_n,target)
```

```python
Classifier.score(inputs_n,target)
```

```python
#a=[temp,max,min,hum,wind ,visib,desc]
array=[290,285,295,99,2.09,10000,6]
a = np.asarray(array).reshape(1,-1)

Classifier.predict(a)
```
